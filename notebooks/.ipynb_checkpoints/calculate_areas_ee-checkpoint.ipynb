# Calculate area
Calculate the area of each fire severity class for all wildfires and prescribed burns from 2005-2020 in the Northern Jarrah Forest, Western Australia (calculate the areas in Table 3 from Dixon et al., 2022) 

```python
import ee
ee.Initialize()
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file(r"C:\Users\1dand\PycharmProjects\njf-fire-sev\data\vector\DBCAFireHistory_2005_2020_gte50ha.shp")
fires = list(gdf['fireID'].unique())
imc = ee.ImageCollection('projects/euc-fire-sw-2022/assets/severity_pred_forested')
```

```python

l = []
for f in fires: 
    print(f)
    try:
        img = imc.filterMetadata('ID','equals',f).first()
        fire_polygons = ee.FeatureCollection(
                'users/danieljdixon1991/NJF/Fire/paper2/data/DBCAFireHistory_2005_2021_all_fires_explode_wID_edited_geometry2')

        fire_poly = ee.Feature(fire_polygons.filterMetadata('fireID', 'equals', f).first())

        area_img = ee.Image.pixelArea().divide(10000)
        img = img.addBands(area_img)

        c1_mask = img.select('f_severity').eq(1)
        c1_area = img.updateMask(c1_mask).select('area').rename('1UB_ha')

        c2_mask = img.select('f_severity').eq(2)
        c2_area = img.updateMask(c2_mask).select('area').rename('2LCS_ha')

        c3_mask = img.select('f_severity').eq(3)
        c3_area = img.updateMask(c3_mask).select('area').rename('3MCS_ha')

        c4_mask = img.select('f_severity').eq(4)
        c4_area = img.updateMask(c4_mask).select('area').rename('4HCS_ha')

        c5_mask = img.select('f_severity').eq(5)
        c5_area = img.updateMask(c5_mask).select('area').rename('5CB_ha')

        img = img.addBands([c1_area, c2_area, c3_area, c4_area, c5_area])

        img = img.select(['1UB_ha', '2LCS_ha', '3MCS_ha', '4HCS_ha', '5CB_ha'])
        area_sum = img.reduceRegion(
            reducer=ee.Reducer.sum(),
            geometry=fire_poly.geometry(),
            scale=30).getInfo()

        area_df = pd.DataFrame([area_sum])
        area_df['f'] = f
        l.append(area_df)
    except:
        pass
```

```python
# add in wf pb and modis yes no
df_dates = pd.read_csv(r"C:\Users\1dand\PycharmProjects\euc-fire-sw-2022\data\date_info\all_fires_dates-start-end.csv")
df_dates['date_adj'].unique()
det = df_dates[df_dates['date_adj'].isin(['modis_1k', 'modis_250m', 'manual'])]
det['modis_det'] = 'modis'
ndet = df_dates[df_dates['date_adj'].isin(['no_adj'])]
ndet['modis_det'] = 'no_modis'
c = pd.concat([det, ndet])
modis_dict = dict(zip(c['ID'], c['modis_det']))
type_dict = dict(zip(c['ID'], c['fih_fire_t']))
```

```python
df2 = pd.concat(l)
df2['modis'] = df2['f'].map(modis_dict)
df2['type'] = df2['f'].map(type_dict)

final_area = df2.groupby(['type','modis'])['1UB_ha','2LCS_ha','3MCS_ha','4HCS_ha','5CB_ha'].sum()#.reset_index()
final_area[['1UB_ha','2LCS_ha','3MCS_ha','4HCS_ha','5CB_ha']] / 100
```
